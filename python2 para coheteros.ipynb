# While

## Reduce

# Matplotlib

Matplotlib es un m√≥dulo de dibujo de gr√°ficas 2D y 3D,  y una de las mas usada por su semejanza a Matlab es pyplot. 

http://matplotlib.sourceforge.net/, ejemplos de como hacer dibujos con Matplotlib.
La documentaci√≥n oficial se encuentra en http://matplotlib.sourceforge.net/contents.html 

Funciones:
figure(num, figsize, dpi, facecolor, edgecolor, frameon) Nos permite crear una nueva figura
    num = numeraci√≥n de la figura, si num = None, las figuras se numeran autom√°ticamente.
    figsize = w, h tuplas en pulgadas. Tama√±o de la figura
    dpi = Resoluci√≥n de la imagen en puntos por pulgada.
    facecolor = Color del rect√°ngulo de la figura.
    edgecolor = Color del per√≠metro de la figura.
    
 subplot(numRows, numCols, plotNum) Permite incluir varias gr√°ficas en una √∫nica figura.  
    frameon = Si es falso, elimina el marco de la figura.
    numRows = N√∫mero de filas
    numCols = N√∫mero de columnas
    plotNum = N√∫mero de gr√°fica
    
 plot(x, y, linestyle, linewidth, marker) Permite incluir varias gr√°ficas en una √∫nica figura. 
    x = Abcisas.
    y = Ordenadas. Tanto x como y pueden ser abcisas tuplas, listas o arrays. La √∫nica condici√≥n es que el tama√±o de ambas debe ser el mismo ya que en caso contrario python nos devolver√° un fallo de tipo dimesi√≥n. Tambi√©n se puede hacer una gr√°fica sin especificar la coordenada x.
    linestyle = color y tipo de dibujar la gr√°fica. Por ejemplo ‚Äòk- -‚Äò
    linewidth = ancho de l√≠nea.
    marker = Marcador.

Tipos Los tipos m√°s comunes son:

‚Äò - -‚Äò l√≠nea a rayas

```python
#Para colocar varias graficas

import numpy as np                  
import matplotlib.pyplot as plt     

# Definimos el periodo de las graficas
periodo = 5

# Definimos el array dimensional
x = np.linspace(0, 15, 2000)
y = np.cos(2*np.pi*(x+2)/periodo+3)  #definimos la funcion

# Creamos la figura
plt.figure()

# Primera grafica
plt.subplot(2,2,1)
plt.plot(x, y,'y')

#segunda grafica
plt.subplot(2,2,2)
plt.plot(x, y,'b')

# Tercera grafica
plt.subplot(2,2,3)
plt.plot(x, y,'m')

# Cuarta grafica
plt.subplot(2,2,4)
plt.plot(x, y,'r')

# Mostramos en pantalla
plt.show()
```

```python
import numpy as np                 
import matplotlib.pyplot as plt     

# Definimos el periodo de la funcion
periodo = 0.8

# Definimos el array dimensional
x = np.linspace(0, 2, 1000)

# Definimos la funcion senoidal
y = np.sin(2*np.pi*x/periodo) #si es coseno = cos

# Creamos la figura
plt.figure()

# Dibujamos  en verde punteado y lo etiquemos como A
plt.plot(x, y, 'g:', linewidth = 3, label = 'A')#label (etiqueta), Linewidth (ancho de linea)

# Mantenemos la misma figura parta la siguiente grafica
plt.hold(True)

# Esta vez dibujamos - y en magenta con etiqueta B
plt.plot(x,-y,'m', linewidth = 2, label = 'B')

# Colocamos una leyenda correspondiente
plt.legend(loc = 3) #cambia la posicion segun el cuadrante de la grafica de forma antihoraria

# Colocamos las etiquetas poniermo en Latex "mu" s√É¬≠mbolo de micras
#fontsize es el tama√±o de la letra de la etiqueta

plt.xlabel(r"$x (\mu m)$", fontsize = 20, color = (1,0,0))
plt.ylabel(r"$y (\mu m)$", fontsize = 20, color = 'red')

# Colocamos el texto correspondiente
#x,y nos dan la posicion del texto dentro del marco
plt.text(x = 1, y = 0.2, s = u'T = 0.08', fontsize = 15)

# Colocamos la grilla
plt.grid(True)
plt.grid(color = '0.3', linestyle = '--', linewidth = 1)

# Colocamos los ejes correspondientes
plt.axis('tight')

# Ponemos el titulo
plt.title('(Dos senos)',fontsize = 15, color = 'b', verticalalignment = 'baseline', horizontalalignment = 'center')

# Guardamos
plt.savefig('Dos senos.png')

# Mostramos en pantalla
plt.show()
```

```python
#Graficando en dos dimensiones

import numpy as np                  
import matplotlib.pyplot as plt     

# Creamos una figura
plt.figure()

# Creamos los arrays dimensionales
x = np.arange(-10, 5, 0.01)
y = np.arange(-5, 5, 0.01)


# Obtenemos las corrdenadas resultantes de esos arrays
X, Y = np.meshgrid(x, y)

# Definimos la grafica cos(x^3 + y^2+2)
fxy = np.cos(X**2+Y**2+2)

# Representamos
plt.imshow(fxy);

plt.colorbar();

# Mostramos en pantalla
plt.show()
```

```python
#Para Graficar en 3D

import numpy as np
import matplotlib.pyplot as plt  
from mpl_toolkits.mplot3d import Axes3D  #nos permite graficar en 3D       
from scipy.misc import imread                   
                              
               
superficie = imread('scua.jpg') #leo la imagen que usare

plt.figure() # Creo una figura

plt.imshow(superficie) # Muestro la imagen en pantalla

# Coloco las etiquetas correspondiente
plt.title('Logo del Semillero')
plt.xlabel('Chicos')
plt.ylabel('Chicas')

logo3d = plt.figure() #creo una figura y la almaceno

ax = logo3d.gca(projection = '3d') #indicamos la representacion en 3d

# Creamos los arrays de la misma dimension que la superficie 
X = np.linspace(-10, 3, superficie.shape[0])
Y = np.linspace(-10, 5, superficie.shape[1])

# Obtenemos las coordenadas a partir de los arrays creados
X, Y = np.meshgrid(X, Y) #meshgrid orienta los ejes al reves

# Defino la funcion a representar
R = np.sqrt(X ** 2 + Y ** 2+3)
Z = np.sin(R)

# Reescalamos de RGB a [0-1]
superficie = superficie.swapaxes(0, 1) / 255. 

ax.plot_surface(X, Y, Z, facecolors = np.flipud(superficie)) #voltear los ejes

# Fijamos la posicion inicial de la grafica
ax.view_init(60, -20) #posicion inicial de la grafica, puedo rotar con esa posicion

plt.title('Despues del rayo')#titulo de la grafica 3d
plt.xlabel('Chicas x') #eje x
plt.ylabel('Chicos') #eje y

# Mostramos en pantalla
plt.show()
```
